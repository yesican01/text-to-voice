```python


!pip install pydub numpy scipy
!apt-get -qq install ffmpeg

from google.colab import files
uploaded = files.upload()

from pydub import AudioSegment, effects
import numpy as np
from scipy.signal import fftconvolve

# Load files
voice = AudioSegment.from_file("male_voice1.mp3")
music = AudioSegment.from_file("male_voice1.mp3")

# 🔊 Normalize voice
voice = effects.normalize(voice)

# 🎶 Lower music volume (so voice is clear)
music = music - 22

# 🔁 Loop music to match voice length
if len(music) < len(voice):
    loops = int(len(voice) / len(music)) + 1
    music = music * loops

music = music[:len(voice)]

# 🌫️ Simple reverb (echo-based)
samples = np.array(voice.get_array_of_samples()).astype(np.float32)

delay_ms = 120
decay = 0.4
delay_samples = int(delay_ms * voice.frame_rate / 1000)

impulse = np.zeros(delay_samples * 2)
impulse[0] = 1.0
impulse[delay_samples] = decay

reverb = fftconvolve(samples, impulse)[:len(samples)]

reverb_audio = voice._spawn(
    reverb.astype(np.int16).tobytes()
)

# 🎧 Mix voice + music
final = music.overlay(reverb_audio)

# 🔊 Final normalize
final = effects.normalize(final)

# 💾 Export
final.export("krishna_final.mp3", format="mp3")



from google.colab import files
files.download("krishna_final.mp3")
```
